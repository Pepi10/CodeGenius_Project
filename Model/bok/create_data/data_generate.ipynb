```python
# 질문 설정
user_prompt1 = "파이썬 개념 관련 기초 질문과 답변, 질문의 핵심 키워드, 예시코드, 공식문서URL을 20개씩 만들어주세요 기존에 알려준 것과 다른 새로운 것으로 알려주세요"
user_prompt2 = "파이썬의 자료형, 연산자 관련 기초 질문과 답변, 질문의 핵심 키워드, 예시코드, 공식문서URL을 20개씩 만들어주세요 기존에 알려준 것과 다른 새로운 것으로 알려주세요"
user_prompt3 = "파이썬의 내장함수 관련 기초 질문과 답변, 질문의 핵심 키워드, 예시코드, 공식문서URL을 20개씩 만들어주세요 기존에 알려준 것과 다른 새로운 것으로 알려주세요"
user_prompt4 = "파이썬의 반복문, 조건문 관련 기초 질문과 답변, 질문의 핵심 키워드, 예시코드, 공식문서URL을 20개씩 만들어주세요 기존에 알려준 것과 다른 새로운 것으로 알려주세요"
user_prompt5 = "파이썬 변수 관련 기초 질문과 답변, 질문의 핵심 키워드, 예시코드, 공식문서URL을 20개씩 만들어주세요 기존에 알려준 것과 다른 새로운 것으로 알려주세요"
user_prompt6 = "파이썬 정규표현식 관련 기초 질문과 답변, 질문의 핵심 키워드, 예시코드, 공식문서URL을 20개씩 만들어주세요 기존에 알려준 것과 다른 새로운 것으로 알려주세요"
user_prompt7 = "파이썬 튜플 관련 기초 질문과 답변, 질문의 핵심 키워드, 예시코드, 공식문서URL을 20개씩 만들어주세요 기존에 알려준 것과 다른 새로운 것으로 알려주세요"
user_prompt8 = "파이썬 리스트 관련 기초 질문과 답변, 질문의 핵심 키워드, 예시코드, 공식문서URL을 20개씩 만들어주세요 기존에 알려준 것과 다른 새로운 것으로 알려주세요"
user_prompt9 = "파이썬 딕셔너리 관련 기초 질문과 답변, 질문의 핵심 키워드, 예시코드, 공식문서URL을 20개씩 만들어주세요 기존에 알려준 것과 다른 새로운 것으로 알려주세요"
user_prompt10 = "파이썬 해쉬셋 관련 기초 질문과 답변, 질문의 핵심 키워드, 예시코드, 공식문서URL을 20개씩 만들어주세요 기존에 알려준 것과 다른 새로운 것으로 알려주세요"
```

```python
# 데이터 생성하는 함수 만들기
import re
import pandas as pd

def generate_and_collect_data(system_prompt, user_prompt, iterations):
    questions = []
    answers = []
    keywords = []
    examples = []
    urls = []

    # 예시로 반복
    for _ in range(iterations):
        res = model.generate_content(system_prompt + user_prompt)
        generated_text = res.text

        # 정규식 패턴 설정 (re.DOTALL 사용)
        pattern = r'질문: (.*?)\n답변: (.*?)\n질문 핵심 키워드: (.*?)\n예시코드: (.*?)\n공식문서 URL: (.*?)\n(?=질문:|$)'

        # 각 텍스트에서 질문과 답변을 추출하여 리스트에 저장
        matches = re.findall(pattern, generated_text, re.DOTALL)
        for match in matches:
            questions.append(match[0])
            answers.append(match[1])
            keywords.append(match[2])
            examples.append(match[3])
            urls.append(match[4])

    # 데이터프레임 생성
    df = pd.DataFrame({'Q': questions, 'A': answers, 'K': keywords, 'E': examples, 'U': urls})
    return df
```

### prompt1

```python
# 최종 데이터를 저장할 데이터프레임 초기화 (컬럼명 설정)
final_df = pd.DataFrame(columns=['Q', 'A', 'K', 'E', 'U'])

# 반복문을 사용하여 여러 번 데이터 생성 및 수집
for i in range(1, 11):
    df = generate_and_collect_data(system_prompt, user_prompt1, iterations=5)
    final_df = pd.concat([final_df, df], ignore_index=True)

# 최종 데이터프레임을 CSV 파일로 저장
final_file_name = 'prompt1_data.csv'
final_df.to_csv(final_file_name, index=False)
print(f"Generated data saved to {final_file_name}")
```
